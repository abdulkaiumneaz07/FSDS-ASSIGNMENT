```python
print("Welcome to the Fast Cash ATM System ")


current_balance = float(input("Please enter your current account balance: "))
requested_amount = float(input("How much money would you like to withdraw today? "))



if requested_amount <= 0:
    print("Invalid Request: Please enter an amount greater than zero.")
elif requested_amount <= current_balance:
    new_balance = current_balance - requested_amount
    print(" Transaction Approved!")
    print(f"Your new balance is: {new_balance}") 
else:
    print("Transaction Denied: Insufficient Funds in your account.")
```

```python
print("Welcome to the Academic Grade Evaluator")

student_score = float(input("Please enter the student's numerical score (0-100): "))



if student_score < 0 or student_score > 100:
    print(" Error: Score must be between 0 and 100. Please try again.")

elif student_score >= 90:
    print(" Congratulations! Your final grade is: A+")
elif student_score >= 80:
    print(" Well done! Your final grade is: A")
elif student_score >= 70:
    print(" Good job! Your final grade is: B")
elif student_score >= 60:
    print(" Satisfactory. Your final grade is: C")
else:
    print(" Needs Improvement. Your final grade is: F (Fail)")
```

```python
print("Welcome to leap year checker:")

input_year = int(input("What is the year you want to check: "))

if input_year % 400 == 0:
    print("Your input year is a leap year")
elif input_year % 4 == 0 and input_year % 100 != 0:
    print("Your input year is a leap year")
else:
    print("Your input year is not a leap year")
```
