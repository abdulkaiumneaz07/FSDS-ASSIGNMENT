# 2. Count Word Frequency in Customer Reviews

### üíº Real-world use: Analyzing reviews or comments in text analytics.


### Task:
Given a review text, count how many times each word appears (ignore case).

### Example:
Input: "This product is good and this service is excellent"

Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
input_sentence = input("Enter a review: ").lower()
input_words = input_sentence.split()

unique_words = []
for word in input_words:
    if word in unique_words:
        pass
    else:
        unique_words.append(word)

counts = []

for i in range(len(unique_words)):
    y = input_words.count(unique_words[i])
    counts.append(y)


pairs = [f"'{unique_words[i]}': {counts[i]}" for i in range(len(unique_words))]
result = '{' + ', '.join(pairs) + '}'

print(result)
```

# 3. Password Strength Checker
üíº Real-world use: Used in user authentication systems.


### Task:
Check if a password is strong ‚Äî must have at least one uppercase, one lowercase, one digit, and one
special character.


### Example:
**Input**: "Hello@123"

**Output**: Strong Password

```python
input_password = input("enter your password:")

upper = []
lower = []
integer = []
extra_symbol = []

for i in input_password:
    if i.isupper():
        upper.append(i)
    elif i.islower():
        lower.append(i)
    elif i.isdigit():
        integer.append(i)
    else:
        extra_symbol.append(i)

if len(upper) != 0 and len(lower) != 0 and len(integer) != 0  and len(extra_symbol) != 0:
    print("Strong Password")
else:
    print("your password is not strong")
```

# 4. Username Generator
üíº Real-world use: Automatically create usernames for users.


### Task:
Take user‚Äôs full name and generate a short username.
Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only).

Example:
**Input**: "Mamun Malitha"
**Output**: mamitha57

```python
User_name = input("Enter your full name: ").lower()
User_words = User_name.split()

first_three = User_words[0][:3]
last_three = User_words[-1][-3:]

username = first_three + last_three

first_word_random = len(User_words[0])
last_three_word_random = len(User_words[-1])

username += str(first_word_random)
username += str(last_three_word_random)

print(username)
```

# 5. SMS Word Counter
üíº Real-world use: Used in messaging apps or social media post limit check.

---

### Task:

Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).

---

### Example:
**Input**: "This is a demo message."

**Output**: Words = 5, Characters = 23, Status = Within Limit

```python
print("Welcome to SMS Word Counter:")


input_sentence = input("Enter your message: ")


word_list = input_sentence.split()
num_words = len(word_list)


Characters = 0
for char in input_sentence:
    Characters += 1

if Characters <= 160:
    status = "Within Limit"
else:
    status = "Not Within Limit"

print(f"Words = {num_words}, Characters = {Characters}, Status = {status}")
```

# 6. Remove Duplicates from Contact List
üíº Real-world use: Data cleaning in CRM or user databases.

---

### Task:
Given a list of names (comma separated), remove duplicates.
Example:
Input: "Rahim, Karim, Rahim, Sultana"
Output: "Rahim, Karim, Sultana"

```python

input_names = input("Enter names separated by commas: ")
name_list = input_names.split(",")

unique_names = []
for name in name_list:
    name = name.strip().capitalize() 
    if name not in unique_names:
        unique_names.append(name)

print(f"Input: {','.join([n.strip().capitalize() for n in name_list])}")
print(f"Output: {', '.join(unique_names)}")
```

# 7. Detect Suspicious Words in Text (Moderation)
üíº Real-world use: Used in chat moderation or email filtering.

----------

### Task:
If any ‚Äúbanned‚Äù words like ['spam', 'scam', 'fake'] are found in a text, print ‚ÄúWarning‚Äù.
Example:
Input: "This is a scam offer"
Output: Warning: Suspicious content detected

```python
banned_words = ["spam", "scam", "fake"]

input_sentence = input("Enter your text: ")

input_word_list = input_sentence.split()

warning_flag = False
for word in input_word_list:
    if word.lower() in banned_words:
        warning_flag = True
        break  

if warning_flag:
    print("Warning: Suspicious content detected")
else:
    print("No suspicious content found")
```

# 8. Invoice Code Formatter
üíº Real-world use: Formatting IDs or codes in finance or ERP systems.

---

### Task:
Given an invoice number like 23, generate code in the format INV00023.

Example:

Input: 23

Output: INV00023

```python

input_code = input("Enter Your code")  
format_str = "INV00000"


format_list = list(format_str)
code_length = len(input_code)

for i in range(1, code_length + 1):
    format_list[-i] = input_code[-i]


output = "".join(format_list)
print(output)
```

# 9. Count Hashtags and Mentions (Social Media Analytics)
üíº Real-world use: Used in analyzing social media posts.

-----


### Task:
Count how many hashtags (#) and mentions (@) appear in a post.
### Example:
 Input: "Check out #AIWorkshop and follow @inceptionbd"

 Output: Hashtags = 1, Mentions = 1

```python
post = input("Enter post: ")

hash_count = 0
mention_count = 0

i = 0
n = len(post)


while i < n:
    if post[i] == '#':
        hash_count += 1
    i += 1

i = 0
while i < n:
    if post[i] == '@':
        if i == 0 or post[i-1] == ' ':
            mention_count += 1
    i += 1

print(f"Hashtags = {hash_count} , Mentions = {mention_count}")
```

# 10. Word Reversal for Data Encryption
üíº Real-world use: Simple text obfuscation or playful encoding.

----

### Task:
Reverse each word in a sentence while keeping word order the same.
### Example:
Input: "AI is fun"

Output: "IA si nuf"

```python
input_sentence = input("Enter your sentence :")
input_sentence_words = input_sentence.split()
reversed_word = []
for i in input_sentence_words:
    j = i[::-1]
    reversed_word.append(j)

print(" ".join(reversed_word))
```
